Conside an unfair probability spinner with $n$ slots. Each slot has a probabilty $P_n \in [0,1]$ and $\sum_{i=0}^{n-1}P_i = 1$. 

Write a function that random generate an index $i$ with the probability $P_i$ from the set $\{0, 1, ..., n\}$.  

```python
import random

def spinner(probabilityDistribution):
    cumulative = []
    total = 0
    for p in probabilityDistribution:
        total += p
        cumulative.append(total)
    
    r = random.random()

    for i, cum_prob in enumerate(cumulative):
        if r < cum_prob:
            return i

P = [0.1, 0.3, 0.2, 0.4]  
index = spinner(P)
print(f"Index: {index}")
```

### Monte Carlo simulation problem
We have a pipe with a circular cross section of **30cm**. The surrounding yard is a square with l = **100cm**.
<br>
The ball uniformly drops inside this surrounding yard.
<br>
Each ball has a **1cm**, and the whole idea behind this is to randomly generate points inside the square yard (uniform distribution), and these points will simulate where the balls will drop. We then check for each point if it falls within the area of the pipe. For simplicity we assume that if the center of the ball is inside of the pipe area then we count it.
We generate random points (x, y) from the range [-50, 50] (because we have a 100cm length in sides). We check if the point falls inside the  pipe, and that will happen if x^2 + y^2 <= R^2 where R is 30cm.

```python
import random
import math

def monte_carlo_simulation(n_trials=100000):
    R = 30      
    l = 100  
    
    count_inside_pipe = 0
    
    for _ in range(n_trials):
        x = random.uniform(-l / 2, l / 2)
        y = random.uniform(-l / 2, l / 2)
        
        if x**2 + y**2 <= R**2:
            count_inside_pipe += 1
    
    probability = count_inside_pipe / n_trials
    return probability

simulated_probability = monte_carlo_simulation()
print(f"Simulated probability: {simulated_probability:.4f}")

R = 30
l = 100
mathematical_probability = (math.pi * R**2) / (l**2)
print(f"Mathematical probability: {mathematical_probability:.4f}")
```
